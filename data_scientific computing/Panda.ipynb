```python
data["Age"].fillna(data["Age"].mean(), inplace=True)
data["Lung Capacity"].fillna(data["Lung Capacity"].mean(), inplace=True)
data["Hospital Visits"].fillna(data["Hospital Visits"].mean(), inplace=True)
data.fillna(method='bfill', inplace=True)
data
```

```python
Q1=data["Hospital Visits"].quantile(0.25)
Q2=data["Hospital Visits"].quantile(0.50)
Q3=data["Hospital Visits"].quantile(0.75)
IQR=Q3-Q1
IQR
```

```python
lowerlimit= Q1-1.5*IQR
upperlimit= Q3+1.5*IQR
lowerlimit,upperlimit
```

```python
data_between = data[data["Hospital Visits"].between(lowerlimit, upperlimit, inclusive='left')]
data_between
```

```python
data_no_outliers = data[data["Hospital Visits"].between(lowerlimit, upperlimit, inclusive='neither')]
data_no_outliers
```

```python
import matplotlib.pyplot as plt

plt.hist(data["Hospital Visits"], bins=15, color='blue', edgecolor='black')
plt.xlabel("Hospital Visits")
plt.ylabel("Frequency")
plt.title("Hospital Visits Histogram")
plt.show()
```

```python
plt.scatter(data["Hospital Visits"], data["Age"])
plt.ylabel("Days")
plt.xlabel("Hospital Visits")
plt.title("Hospital Visits vs Age")
plt.show()
```
