```python
from tabulate import tabulate
table_data = [
    [
        match['match_id'],
        match['match_date'],
        match['match_hometeam_name'],
        match['match_hometeam_score'],
        match['match_awayteam_name'],
        match['match_awayteam_score'], 
        match['match_status'],
        match['match_time'],
        match['match_hometeam_system'],
        match['match_awayteam_system'],
    ]
    for match in data
]

headers = ["Match ID", "Date", "Home Team", "Home Score", "Away Team", "Away Score", "Status", "Time", "Home HT Score", "Away HT Score", "Home Extra Score", "Away Extra Score", "Home Penalty Score", "Away Penalty Score", "Home System", "Away System"]

print(tabulate(table_data, headers=headers, tablefmt="grid"))
```

```python
# Calculate total goals scored by each team
team_goals = {}

for match in data:
    home_team = match['match_hometeam_name']
    away_team = match['match_awayteam_name']
    home_goals = int(match['match_hometeam_score'])
    away_goals = int(match['match_awayteam_score'])
    
    team_goals[home_team] = team_goals.get(home_team, 0) + home_goals
    team_goals[away_team] = team_goals.get(away_team, 0) + away_goals

print(json.dumps(team_goals, indent=4))
team_goal_scorers = {}

for match in data:
    for scorer in match.get('goalscorer', []):
        team = match['match_awayteam_name'] if scorer['home_scorer'] == '' else match['match_hometeam_name']
        player = scorer['away_scorer'] if scorer['home_scorer'] == '' else scorer['home_scorer']
        if team not in team_goal_scorers:
            team_goal_scorers[team] = []
        team_goal_scorers[team].append(player)

print(json.dumps(team_goal_scorers, indent=4))
```

```python
from collections import defaultdict

# Group matches by teams and calculate total goals
team_stats = defaultdict(lambda: {"matches_played": 0, "total_goals": 0})

for match in data:
    home_team = match['match_hometeam_name']
    away_team = match['match_awayteam_name']
    home_goals = int(match['match_hometeam_score'])
    away_goals = int(match['match_awayteam_score'])
    
    team_stats[home_team]["matches_played"] += 1
    team_stats[home_team]["total_goals"] += home_goals
    
    team_stats[away_team]["matches_played"] += 1
    team_stats[away_team]["total_goals"] += away_goals

print(json.dumps(team_stats, indent=4))
```

```python
from collections import defaultdict
league_stats = defaultdict(lambda: {"matches_played": 0, "total_goals": 0})

for match in data:
    league_name = match['league_name']
    home_goals = int(match['match_hometeam_score'])
    away_goals = int(match['match_awayteam_score'])
    
    league_stats[league_name]["matches_played"] += 1
    league_stats[league_name]["total_goals"] += home_goals + away_goals

print(json.dumps(league_stats, indent=4))
```

```python
from collections import Counter

# Group matches by their status and count occurrences
match_status_counts = Counter(match['match_status'] for match in data)

print(json.dumps(match_status_counts, indent=4))
```
